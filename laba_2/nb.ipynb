> розглянемо послідовні типи даних
- list - через те що він mutable його можна змінювати коли захочеш;
- tuple - а він є immutable і змінювати його не можна;

````python
l = [1, "2", 3]
ll = list([1, "2", 3])
print("Маємо список, який складається з елементів:","\nl=",l,"\nll=",ll)

l.append(["New", "List"])
print("Тепер список буде:", l)
l.insert(4, "Це буде в кінці")
print("Після вставки елементу на позицію перед 4 будемо мати:", l)

# Видаляємо передостанній елемент та додаємо нові значення
removed_element = l.pop(-2)
removed_element.extend([-1, "тепер це останній елемент"])

print("Елементи масиву зараз:", l)
print("Те що ми видалили з списку та модифікували:", removed_element)
````

> Спитались в ChatGPT чому так, ось пояснення:
Цей код має на меті продемонструвати різні операції зі списками в мові програмування Python. Давайте розглянемо його крок за кроком:

1. **Створення списків:**
   ```python
   l = [1, "2", 3]
   ll = list([1, "2", 3])
   ```
   У цих рядках створюються два однакових списки `l` і `ll`, які складаються з трьох елементів: ціле число 1, рядок "2" і ціле число 3.

2. **Вивід початкових списків:**
   ```python
   print("Маємо список, який складається з елементів:","\nl=",l,"\nll=",ll)
   ```
   Цей рядок виводить початкові значення обох списків.

3. **Додавання елементу вкінці списку:**
   ```python
   l.append(["New", "List"])
   ```
   В цьому рядку до списку `l` додається новий елемент - список `["New", "List"]`.

4. **Вставка елементу на певну позицію:**
   ```python
   l.insert(4, "Це буде в кінці")
   ```
   Ця операція вставляє рядок "Це буде в кінці" на позицію після четвертого елемента у списку `l`.

5. **Видалення елемента за індексом та модифікація його значення:**
   ```python
   removed_element = l.pop(-2)
   removed_element.extend([-1, "тепер це останній елемент"])
   ```
   Цей рядок видаляє передостанній елемент зі списку `l` за допомогою `pop()` і зберігає його значення у змінній `removed_element`. Потім цей видалений елемент, який є списком, розширюється новими значеннями - числом -1 і рядком "тепер це останній елемент".

6. **Вивід змінених списків та видаленого елемента:**
   ```python
   print("Елементи масиву зараз:", l)
   print("Те що ми видалили з списку та модифікували:", removed_element)
   ```
   Ці рядки виводять змінений список `l` та вміст змінної `removed_element`.

Результат виконання цього коду буде таким:

```
Маємо список, який складається з елементів: 
l= [1, '2', 3, ['New', 'List']] 
ll= [1, '2', 3]
Тепер список буде: [1, '2', 3, ['New', 'List']]
Після вставки елементу на позицію перед 4 будемо мати: [1, '2', 3, ['New', 'List'], 'Це буде в кінці']
Елементи масиву зараз: [1, '2', 3, ['New', 'List'], 'Це буде в кінці']
Те що ми видалили з списку та модифікували: ['New', 'List', -1, 'тепер це останній елемент']
```

> знайомство з Set (набір невпорядкованих, унікальних даних)

````python
s = {1, '1', 2, 1, "1", 3, 5, 7, "Це кінець"}

print("Навіть якщо ми задамо повторювані елементи, всеодно буде:",s)
print("Дані стали невпорядкованими, витягнути елемент по його індексу буде давати помилку, а всі дублікати були видалені з Set.")

s[1]
````

````python
print("Видалеми елемент",s.pop())
s.add("2")
print("Тепер Set  має вигляд:",s)
````

> знайомство з словниками

````python
a = "Це значення змінної а"
l = [1, "List", 2]
k = "k_k"
d = {"key": "value", 1: 2, "Це є ключ": "Це його значення", "a": a, "l": l, k: k*5}
print("Утворений словник буде мати вигляд:\n", d)

print("У списках ми доступаємось по індексах, беремо індекс 1:",l[1],
      "\nВ словниках для доступу до елемента використовуємо його ключ, беремо ключ 'Це є ключ':",d["Це є ключ"])

k2 = 'k_k'
print("Навіть якщо використати нову змінну, але в якій буде записано правильний ключ, ми зможемо доступитись до значення з словника.",
      "\nЗначення буде:",d[k2])

d.pop(k2)
print("Після видалення буде:",d)
print("Можемо переглянути всі наявні в словнику ключі:", d.keys())
print("Можемо переглянути всі наявні в словнику значення:", d.values())

dd = {"Прізвище": ["Пупкін", "Персона_2"], "імя": ["Іван", "Ілон"]}
print("Приклад складних типів даних можу бути таким", dd,
      "\nДоступатись до нього ми можемо ось так:", dd["Прізвище"][0])
````
