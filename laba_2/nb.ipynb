### Основи Python
> Основні типи даних
1. поняття змінних у Python

```python
a = "" # це є змінна `a`, їй присвоєна пуста стрічка
print("Значення змінної: ", a) # оскільки a є присвоєно пусту стрічну, то нічого не виведеться

a = 'Тут вже буде стрічка' # переприсвоїли значення змінної a
print("Значення змінної: ", a, "\nЦя змінна буде типу: ", type(a))

a = str("Це значення буде в обєкті текстового типу")
print("Значення змінної: ", a, "\nТут ми вже явно задаємо тип змінно. який буде: ", type(a))

# Спробуємо попацювати з змінною яка є обєктом стрічкового/текстового типу
print("Перевіряємо чи дана стрічка має цифри: ", a.isdigit())
```

> Конкатенація стрічок, множення та додавання стрічок

```python
print("Значення змінної а є цифрою?", a.isdigit())
b = "Це є продовження стрічки"
print("Значення змінної b є цифрою?", b.isdigit())
print("Що ж буде якщо ми додамо дві стрічки які не є цифрами?")
c = a + " " + b
print(c)

print("А що буде якщо ми помножимо стрічку на число?")

d = "Це будем множити на число"
print(d * 5)

e = d.__add__( "Це ми додали до d")
print("Використовуючи ООП та методи __add__ ми отримаємо результат: ", e)
```
